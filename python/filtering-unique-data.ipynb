```python
df_before=df['before'].dropna()
df_after=df['after']
print('before:', df_before.shape)
print('after:', df_after.shape)
```

```python
df_before.head()
```

#### set( ) 연산으로 unique한 값만 뽑기

```python
# before에는 있고, after에는 없는것
only_before=pd.DataFrame(set(df_before) - set(df_after),columns=['qm'])
only_before.shape
```

```python
# after에는 있고, before에는 없는것
only_after=pd.DataFrame(set(df_after) - set(df_before),columns=['qm'])
only_after.shape
```
